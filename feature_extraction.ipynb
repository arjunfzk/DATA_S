```python
survivedQ = titanic[titanic.Embarked == 'Q'][titanic.Survived == 1].shape[0]
survivedC = titanic[titanic.Embarked == 'C'][titanic.Survived == 1].shape[0]
survivedS = titanic[titanic.Embarked == 'S'][titanic.Survived == 1].shape[0]
usurvivedQ = titanic[titanic.Embarked == 'Q'][titanic.Survived == 0].shape[0]
usurvivedC = titanic[titanic.Embarked == 'C'][titanic.Survived == 0].shape[0]
usurvivedS = titanic[titanic.Embarked == 'S'][titanic.Survived == 0].shape[0]
print(survivedQ,survivedC,survivedS)
print(usurvivedQ,usurvivedC,usurvivedS)
```

```python
def getNumber(str):
    if str=="male":
        return 1
    else:
        return 2
titanic["gender"]=titanic["Sex"].apply(getNumber)
#We have created a new column called "gender" and 
#filling it with values 1 ,2 based on the values of sex column
titanic.head()
```

```python
del titanic["Sex"]
titanic.head()
```

```python
#finding mean survived age
meanS= titanic[titanic.Survived==1].Age.mean()
meanS
```

```python
meanUS= titanic[titanic.Survived==0].Age.mean()
meanUS
```

```python
titanic["age"]=np.where(pd.isnull(titanic.Age) & titanic["Survived"]==1  ,meanS, titanic["Age"])
titanic.head()
```

```python
titanic.age.fillna(meanUS,inplace=True)
titanic.isnull().sum()
```

```python
del titanic["Age"]
titanic.head()
```

```python
titanic.rename(columns={'age':'Age'}, inplace=True)
titanic.rename(columns={'gender':'Sex'}, inplace=True)
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
del titanic['Cabin']
titanic.dropna(inplace=True)
titanic.head()
```

```python
titanic['FamilySize'] = titanic.SibSp.values + titanic.Parch.values + 1
titanic.head()
```

```python
titanic['AgeBand'] = pd.cut(titanic['Age'], 5)
titanic[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
    
titanic.loc[ titanic['Age'] <= 16, 'Age'] = 0
titanic.loc[(titanic['Age'] > 16) & (titanic['Age'] <= 32), 'Age'] = 1
titanic.loc[(titanic['Age'] > 32) & (titanic['Age'] <= 48), 'Age'] = 2
titanic.loc[(titanic['Age'] > 48) & (titanic['Age'] <= 64), 'Age'] = 3
titanic.loc[ titanic['Age'] > 64, 'Age']
titanic.head()
```

```python
titanic['IsAlone'] = 0
titanic.loc[titanic['FamilySize'] == 1, 'IsAlone'] = 1
titanic.head()
```

```python
x=titanic[["SibSp", "Survived"]].groupby(['SibSp'], as_index=False).mean().sort_values(by='Survived', ascending=False)
x1=titanic[["IsAlone", "Survived"]].groupby(['IsAlone'], as_index=False).mean().sort_values(by='Survived', ascending=False)
x2=titanic[["FamilySize", "Survived"]].groupby(['FamilySize'], as_index=False).mean().sort_values(by='Survived', ascending=False)
x3=titanic[["Parch", "Survived"]].groupby(['Parch'], as_index=False).mean().sort_values(by='Survived', ascending=False)
print(x,x1,x2,x3,sep="\n")
```
